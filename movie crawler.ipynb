```python
import csv
import requests
from bs4 import BeautifulSoup

infos = []
for page1 in range(1, 3):
    url = 'https://movies.yahoo.com.tw/moviegenre_result.html?genre_id=%s' % page1
    for page2 in range(1, 3):
        url += "&page=%s" % page2
        r = requests.get(url)
        soup = BeautifulSoup(r.text)

        for d in soup.find_all('div', class_="release_info"):
            info = {}
            url = d.find('div', class_='release_text').find('span')['data-url']
            r = requests.get(url)
            soup = BeautifulSoup(r.text)
            for j in soup.find_all('div', class_="movie_intro_info _c"):

                if(info.get("中文名稱") == j.find('h1').text.strip()):
                    continue

                info['中文名稱'] = j.find('h1').text.strip()
                info['英文名稱'] = j.find('h3').text.strip()
                info['類型'] = j.find('div', class_= 'level_name_box').text.replace("\n", "").replace(" ", "").strip()
            for k in soup.find_all('div', class_="gray_infobox_inner"):
                info['劇情介紹'] = k.find('span', id="story").text.strip()
            
            infos.append(info)



infos

with open('info.csv', 'w', encoding='utf-8', newline='') as csv_file:

    csv_writer = csv.writer(csv_file)
    csv_writer.writerow(['中文名稱', '英文名稱', '分類', '劇情介紹'])

    for item in infos:
        name = item['中文名稱']
        english_name = item['英文名稱']
        m_class = item['類型']
        content = item['劇情介紹']

        csv_writer.writerow([name, english_name, m_class ,content])
        print('{}({}) 分類：{} 劇情介紹: {} '.format(name, english_name, m_class,content))
```
